Klasyczny oscylator harmoniczny

```python
m, k, x_0, xdot_0, omega_0, t = symbols('m, k, x_0, xdot_0, omega_0, t')
x = Function('x')
```

```python
sol = dsolve(Derivative(x(t), t, 2) + omega_0**2 * x(t))
sol
```

```python
ics = [Eq(sol.args[1].subs(t, 0), x_0), Eq(sol.args[1].diff(t).subs(t, 0), xdot_0)]
ics
```

```python
solved_ics = solve(ics)
solved_ics
```

```python
full_sol = sol.subs(solved_ics[0])
full_sol
```

```python
case1 = simplify(full_sol.subs({x_0:1, xdot_0:0, omega_0:2}))
case1
```

```python
plot(case1.rhs)
```

```python
case2 = simplify(full_sol.subs({x_0:0, xdot_0:1, omega_0:2}))
case2
```

```python
plot(case2.rhs)
```
